```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats
import statsmodels.stats
```

```python
data = pd.read_csv('C:\\Users\\juhic\\OneDrive\\Desktop\\master_dataset.csv')
data.drop(axis = 1, columns = 'Unnamed: 0', inplace = True)

data = data.astype({'is_volume':'category',
                    'author_sex':'category',
                    'author_exp':'category',
                    'book_size':'category',
                    'genre':'category',
                    'genre_category':'category'})
data.info()
```

```python
cols = ['rating_count','genre_category','book_size','is_volume',
        'page_count','author_work_count','author_fan_count','author_review_count',
        'author_sex','author_exp','author_rating_count','author_avg_rating']
df = data[cols]
df.head()
```

```python
# One Hot Encoding:

df['fiction'] = [1 if i == 'fiction' else 0 for i in df['genre_category']]
df['non_fiction'] = [1 if i == 'non-fiction' else 0 for i in df['genre_category']]

df['light'] = [1 if i == 'light' else 0 for i in df['book_size']]
df['bulky'] = [1 if i == 'bulky' else 0 for i in df['book_size']]

df['is_volume'] = [1 if i == 'yes' else 0 for i in df['is_volume']]


df['newbie_author'] = [1 if i == 'newbie' else 0 for i in df['author_exp']]
df['experienced_author'] = [1 if i == 'experienced' else 0 for i in df['author_exp']]

df['male_author'] = [1 if i == 'male' else 0 for i in df['author_sex']]

df.head()
```

```python
import statsmodels.api as sm

y = df['rating_count']                         

predictor_cols = ['fiction','non_fiction','light','bulky','is_volume',
                  'male_author','newbie_author','experienced_author',
                  'author_rating_count','author_avg_rating']
x = df[predictor_cols]
x = sm.add_constant(x)

model1 = sm.OLS(y,x).fit()
print(model1.summary())
```

```python
df['author_avg_rating_squared'] = df['author_avg_rating'].apply(lambda x: pow(x,10))

y = df['rating_count']                         

predictor_cols2 = ['fiction','non_fiction','light','bulky','is_volume',
                  'male_author','newbie_author','experienced_author',
                  'author_rating_count','author_avg_rating','author_avg_rating_squared']
x2 = df[predictor_cols2]
x2 = sm.add_constant(x2)

model2 = sm.OLS(y,x2).fit()
print(model2.summary())
```

```python
df['author_rating_count_squared'] = df['author_rating_count'].apply(lambda x: pow(x,2))

y = df['rating_count']                         

predictor_cols3 = ['fiction','non_fiction','light','bulky','is_volume',
                  'male_author','newbie_author','experienced_author',
                  'author_rating_count','author_avg_rating','author_rating_count_squared']
x3 = df[predictor_cols3]
x3 = sm.add_constant(x3)

model3 = sm.OLS(y,x3).fit()
print(model3.summary())
```
